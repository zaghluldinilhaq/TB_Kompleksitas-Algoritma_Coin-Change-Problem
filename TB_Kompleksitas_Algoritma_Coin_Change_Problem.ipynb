```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def coin_change(coins, amount):
    coins.sort(reverse=True)
    count = 0
    coin_count = []
    steps = []

    for coin in coins:
        while amount >= coin:
            amount -= coin
            count += 1
            coin_count.append(coin)
            steps.append((coin, amount))

    return count, coin_count, steps

def visualize_steps(steps, title):
    fig, ax = plt.subplots()
    coins_used, remaining_amounts = zip(*steps)
    step_numbers = list(range(1, len(steps) + 1))

    ax.plot(step_numbers, remaining_amounts, marker='o', linestyle='-', color='b')
    for i, txt in enumerate(coins_used):
        ax.annotate(f"Koin {txt}", (step_numbers[i], remaining_amounts[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=9, color='red')

    ax.set_xlabel("Langkah", fontsize=12)
    ax.set_ylabel("Jumlah Uang Tersisa (Rp)", fontsize=12)
    ax.set_title(title, fontsize=14)
    ax.grid(True, linestyle='--', alpha=0.6)

    # Membuat anotasi untuk koin yang digunakan
    legend_patches = [mpatches.Patch(color='red', label='Koin yang digunakan')]
    ax.legend(handles=legend_patches, loc='upper right')

    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()

# Contoh penggunaan
denominations = [100, 200, 500, 1000, 2000]

# Kasus 1: Kembalian Rp 5700
change_amount_1 = 5700
total_coins_1, coins_used_1, steps_1 = coin_change(denominations, change_amount_1)
print("Jumlah koin yang dibutuhkan (Rp 5700):", total_coins_1)
print("Koin yang digunakan (Rp 5700):", coins_used_1)
visualize_steps(steps_1, "Proses Pengembalian Koin (Rp 5700)")

# Kasus 2: Kembalian Rp 8300
change_amount_2 = 8300
total_coins_2, coins_used_2, steps_2 = coin_change(denominations, change_amount_2)
print("\nJumlah koin yang dibutuhkan (Rp 8300):", total_coins_2)
print("Koin yang digunakan (Rp 8300):", coins_used_2)
visualize_steps(steps_2, "Proses Pengembalian Koin (Rp 8300)")

# Kasus 3: Kembalian Rp 12300
change_amount_3 = 12300
total_coins_3, coins_used_3, steps_3 = coin_change(denominations, change_amount_3)
print("\nJumlah koin yang dibutuhkan (Rp 12300):", total_coins_3)
print("Koin yang digunakan (Rp 12300):", coins_used_3)
visualize_steps(steps_3, "Proses Pengembalian Koin (Rp 12300)")
```
